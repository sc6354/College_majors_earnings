## Sec. 3.1: Majors with most Males and Females

```python
#create a new column that shows the difference between male and female graduates in each major
#we will use this new column to sort the data

df["delta_graduates"] = df["Men"] - df["Women"]

#print
print(df.groupby(["Major_category"])["delta_graduates"].sum().sort_values())
```

```python
df.groupby('Major_category').sum().sort_values(by=['delta_graduates'] ,ascending=False).plot.barh(
    y=['Men','Women'], figsize =(20,10))
sns.despine(left=True, bottom=True)
plt.xlabel('Total Number of Gradudate')
plt.ylabel('Majors')
#plt.legend("gender_majority")
plt.title("Distribution of Men and Women in each Major Categories", weight='bold').set_fontsize('25')
```

#### Top 3 majors that are predominantly female: Education, Health, and Psychology & Social Work. (in order)

#### Top 3 majors that are predominantly male: Engineering, Computers & Mathematics, and Business. (in order)

## Sec 3.2: Median Salary and Unemployment rate

```python
fig, axes = plt.subplots(1, 2, figsize= (16,7), sharey=False)
sns.boxplot(ax=axes[0], data = df['Median'])
axes[0].set_title('Distribution of Median Salary ($)')
sns.boxplot(ax=axes[1], data = df['Unemployment_rate'])
axes[1].set_title('Distribution of Unemployment Rate (%)')
```

#### The medain salary hovers around `$30,000 to $40,000` and the common unemployment rates are between `5.5% - 6.25%`

## Sec. 3.3: Top 5 majors with the highest 75th percentile of earnings

```python
major_highest_P75th = df.groupby(["Major"])["P75th"].mean().sort_values(ascending=False).head()
print(major_highest_P75th)

fig, ax =plt.subplots(1,1,figsize=(15,6))
plt.title("Top 5 majors with the highest 75th percentile of earnings", weight='bold').set_fontsize('25')
sns.barplot(x=major_highest_P75th, y=major_highest_P75th.index, data=df, ci=None)
ax.set(xlim=(0, 140000), xlabel="Salary at the 75th Percentile")
#sns.despine(right=True,top=True)
```

## Sec. 3.4: Job Security 
 - Top 5 majors with the most full time employees

```python
major_most_fulltime = df.groupby(["Major"])["Full_time_year_round"].mean().sort_values(ascending=False).head()
print(major_most_fulltime)

fig, ax =plt.subplots(1,1,figsize=(15,6))
plt.title("Top 5 majors with the most full time employees", weight='bold').set_fontsize('25')
sns.barplot(x=major_most_fulltime, y=major_most_fulltime.index, data=df, ci=None)

#sns.despine(left=True, bottom=True)
ax.set(xlim=(0, 225000), xlabel="Mean Number of Full Time Employee")
```

Graduates who studied `Business Management and Administration`, `Pyschology`, `General Business`, `Marketing and Marketing Research`, or `Accounting` had greater job stability and ease of getting a job compared to graduates from other majaors.

# Sec. 4: Examining the relationship between majors, salary, and gender
***

## Sec.4.1 : Do more popular majors make more money?

- The two variables examined here are 'Total' and 'Median Salary'
- Total represents the number of people with this major
- Median Salary is the median salary of full-time, year-round workers

```python
fig, ax =plt.subplots(1,1, figsize =(10,5))
sns.scatterplot(data = df, x='Total', y='Median', hue ='Median')
sns.despine(left=True, bottom=True)
plt.title('Total vs Median', weight='bold').set_fontsize('20')
plt.legend(bbox_to_anchor=(1.55, 0.5), title= 'Median')
plt.show()
```

#### Based on the plot, the answer is no. There is no correlation between Median and Total. We can see that there is a high variance near the 0 points, but as Total increases the variance diminishes. In other words, there is no correlation between the popularity of the major, and money

## Sec.4.2 : Do majors with majority female graduates make more money?

- The two variables examined here are 'ShareWomen' and 'Median Salary'
- ShareWomen represents women graduates as a proportion of Total for a specific major 

```python
fig, ax =plt.subplots(1,1, figsize =(10,5))
sns.scatterplot(data = df, x='ShareWomen', y='Median', hue ='Median')
sns.despine(left=True, bottom=True)
plt.title('Median vs. ShareWomen', weight='bold').set_fontsize('15')
plt.legend(bbox_to_anchor=(1.55, 0.5), title= 'Median')
plt.show()
```

#### Based on the plot, there is a weak negative correlation between Median and ShareWomen. We can see that as Sharewomen increases, the Median decreases.

#### In other words, the students that majored in subjects that had female majority tend to make less money compared to their peers. To give a better idea of how much less, the next subsection divides the data into three bins and plots the average median salary of each of those bins.

## Sec. 4.3: Taking a Deeper diving on Median and ShareWomen

```python
#Analyze ShareWomen and Median using groupby()

median_sharewomen = df.groupby(["ShareWomen"])["Median"].mean().sort_values(ascending=False)
print(median_sharewomen)
```

#### The result suggest that ShareWomen with low values tend to get higher median earnings, and vice versa.
#### However, to be more confident in our analysis , let's group ShareWomen into 3 bins, and see what is the average Median of each group.

```python
#splitting ShareWomen into 3 groups
sharewomen_grouped = df["ShareWomen"].value_counts(bins = 3).sort_index(ascending= False)
sharewomen_grouped
```

```python
bins = [-0.0019690000000000003, 0.323, 0.646, 0.969] #using the bins from the previous result

median_sharewomen_grouped = df.groupby(pd.cut(df["ShareWomen"], bins))["Median"].mean().sort_values(ascending= False)
print(median_sharewomen_grouped)

fig, ax =plt.subplots(1,1, figsize =(10,5))

plt.xlabel('ShareWomen')
plt.ylabel('Median')
sns.barplot(x=sorted(median_sharewomen_grouped.index), y=median_sharewomen_grouped, data=df, ci=None)
plt.title('Median vs ShareWomen (bins = 3)', weight='bold').set_fontsize('20')
#sns.despine(left=True, bottom=True)
```

#### This plot further validates that students that majored in subjects with female majority on average made less money.

#### Summary:

#### - Do more popular majors make more money?
    Answer: No. There is correlation between the popularity of the major and salary.

#### - Do majors with majority female graduate make more money?
    Answer: The opposite is true. The students that majored in subjects that were female majority on average make less money.
